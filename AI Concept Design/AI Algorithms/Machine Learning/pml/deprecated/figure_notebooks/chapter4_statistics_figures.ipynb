## Figure 4.11:<a name='4.11'></a> <a name='sequentialCoinToss'></a> 


 Illustration of sequential Bayesian updating for the beta-Bernoulli model. Each colored box represents the predicted distribution $p(x_t| \bm h  _t)$, where $ \bm h  _t=(N_ 1,t ,N_ 0,t )$ is the sufficient statistic derived from history of observations up until time $t$, namely the total number of heads and tails. The probability of heads (blue bar) is given by $p(x_t=1| \bm h  _t) = (N_ t,1  + 1)/(t+2)$, assuming we start with a uniform $\mathrm Beta (\theta |1,1)$ prior.
